```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load your dataset
data = pd.read_csv('Salaries.csv')
data.head()
```

```python
data.columns
```

##Basic Data Exploration


```python
#Number of rows and columns
num_rows, num_columns = data.shape
print(f"Number of rows in data set is : {num_rows}")
print(f"Number of columns in data set is : {num_columns}")
```

```python
#data types of columns
data_types = data.dtypes
print("\nData types of each column are:")
print(data_types)
```

```python
# Check missing values in each column
missing_values = data.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
#show dataset columns with more details
data.info()
```

##Descriptive Statistics


```python
# Calculate basic statistics for ALL columns
basic_stats = data.describe()
print(basic_stats)
```

```python
# Calculate range of salaries only
salary_range = data['TotalPay'].max() - data['TotalPay'].min()
print(salary_range)
```

```python
# Calculate standard deviation of salaries only
std_dev = data['TotalPay'].std()
print(std_dev)
```

##Data Cleaning

```python
data.info()
```

```python
#delete non_value columns
#becuase these are empety columns
data.drop(['Notes','Status'],axis='columns',inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#handel null_values in numeric columns with fill it with the mean value of columns
data['BasePay'] = data['BasePay'].fillna(data['BasePay'].mean())
data['OvertimePay'] = data['OvertimePay'].fillna(data['OvertimePay'].mean())
data['OtherPay'] = data['OtherPay'].fillna(data['OtherPay'].mean())
```

```python
#handel null_values of Benefits column with fill it with 0 value
#becuase don't effect on the salary values
data['Benefits'] = data['Benefits'].fillna(0)
```

```python
#check null values in columns
data.isnull().sum()
```

##Basic Data Visualization

```python
# Visualization of salary distribution using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalPay'], bins=20, kde=True, color='skyblue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization of the proportion of employees in different departments using a pie chart
department_counts = data['Agency'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Agencies')
plt.show()
```

```python
#this chart show that all employees from San Francisco
```

##Grouped Analysis

```python
employee_data = pd.DataFrame(data)
```

```python
grouped_data = employee_data.groupby(['JobTitle', 'Year'])
average_salary_by_group = grouped_data['TotalPay'].mean()
```

```python
average_salary_by_group.unstack().plot(kind='bar', figsize=(12, 6), title='Average Salary by Job Title and Year')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.show()
```

##Simple Correlation Analysis

```python
correlation = employee_data['TotalPay'].corr(employee_data['Benefits'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(employee_data['TotalPay'], employee_data['Benefits'])
plt.title('Scatter Plot: TotalPay vs Benefits')
plt.xlabel('TotalPay')
plt.ylabel('Benefits')
plt.grid(True)
plt.show()
```

```python
correlation = employee_data['TotalPay'].corr(employee_data['Year'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(employee_data['TotalPay'], employee_data['Year'])
plt.title('Scatter Plot: TotalPay vs Years')
plt.xlabel('TotalPay')
plt.ylabel('Years')
plt.grid(True)
plt.show()
```
